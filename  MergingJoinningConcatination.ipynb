```python
import pandas as pd
```

```python
#merging two dataset
```

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['Nida', 'Maha', 'Sama', 'Sana', 'Anum'],
    'department': ['IT','HR', 'FINANCE', 'IT', 'HR']
})
salaries = pd.DataFrame({
    'employee_id': [1, 2, 4, 6, 7],
    'salary': [100000, 50000, 40000, 60000,70000],
    'bonus': [20000, 10000, 5000, 7000, 9000]
}
)
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries)
```

```python
pd.merge(employees, salaries, on = 'employee_id',how='inner')
```

```python
pd.merge(employees, salaries, on = 'employee_id',how='outer')
```

```python
pd.merge(employees, salaries, on = 'employee_id',how='left')
```

```python
pd.merge(employees, salaries, on = 'employee_id',how='right')
```

```python
#concatinating two datasets
```

```python
df1 = pd.DataFrame({
    'A' : ['A0', 'A1', 'A2'],
    'B' : ['B0', 'B1', 'B2'], 
    'C' : ['C0', 'C1', 'C2']
})
df2 = pd.DataFrame({
    'A' : ['A3', 'A4', 'A5'],
    'B' : ['B3', 'B4', 'B5'], 
    'C' : ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis =1)
```

```python
#joining two dataset
```

```python
df1 = pd.DataFrame({
     'Name': [' Alice', 'Bob', 'Charlie']
},
     index = [1, 2, 3])
df2 = pd.DataFrame({
    'Score': [85, 90, 75]
},
     index=[2, 3, 4] )
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='outer')
```

```python
df2.join(df1)
```
